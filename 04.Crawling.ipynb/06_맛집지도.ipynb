## 식신에서 찾은 맛집 -> 지도 위에 표시하기

```python
import pandas as pd 
import folium
```

```python
df = pd.read_csv('맛집2.tsv', sep='\t')
df.head()
```

### 배경 지도 그리기

```python
map = folium.Map(location=[df.lat.mean(), df.lng.mean()], zoom_start=13)     # len - 위도, lan - 경도
map
```

```python
map = folium.Map(location=[df.lat.mean(), df.lng.mean()], zoom_start=15)
for i in df.index:
    folium.Marker(
    location=[df.lat[i], df.lng[i]],
    popup=df['상호'][i]
).add_to(map)

map
```

```python
map = folium.Map(location=[df.lat.mean(), df.lng.mean()], zoom_start=15)
for i in df.index:
    folium.Marker(
    location=[df.lat[i], df.lng[i]],
    popup=df['상호'][i]
).add_to(map)

title_html = '<h3 align="center" style="font-size:20px"><b>산격동 맛집 - 식신 추천</b></h3>'
map.get_root().html.add_child(folium.Element(title_html))
map
```

```python
map = folium.Map(location=[df.lat.mean(), df.lng.mean()], tiles="Stamen Toner", zoom_start=15)
for i in df.index:
    folium.Circle(
        radius=30,
        location=[df.lat[i], df.lng[i]],
        popup=df['상호'][i],
        tooltip=df['메뉴'][i],
        color="blue",
        fill=False,
    ).add_to(map)

title_html = '<h3 align="center" style="font-size:20px"><b>산격동 맛집 - 식신 추천</b></h3>'   
map.get_root().html.add_child(folium.Element(title_html))
map
```

```python
# map을 html로 save하기
map.save('맛집 지도.html')
```
