### 다양한 그래프 그리기

```python
import matplotlib as mpl
import matplotlib.pyplot as plt 
%matplotlib inline

mpl.rcParams['axes.unicode_minus'] = False # minus 표시
plt.rcParams["font.family"] = 'Malgun Gothic'
```

1. Line Plot

```python
x = [0, 1, 2, 3]        # 리스트, 넘파이 어레이 가능, 판다스 시리즈
y = [0, 1, 4, 9]
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, color='m',marker='o', linestyle='--')
plt.grid()
plt.show()
```

```python
import numpy as np
X = np.arange(4)
Y1 = X * X 
Y2 = 1.2 * X * X 
Y3 = 0.8 * X * X 
plt.plot(X, Y1)
plt.plot(X, Y2, color='m',marker='o', linestyle='--')
plt.plot(X, Y3, 'k^:')
plt.grid()
plt.show()
```

```python
import numpy as np
X = np.arange(4)
Y1 = X * X 
Y2 = 1.2 * X * X 
Y3 = 0.8 * X * X 
plt.plot(X, Y1, label='Y1')
plt.plot(X, Y2, color='m',marker='o', linestyle='--', label='Y2')
plt.plot(X, Y3, 'k^:', label='Y3')
plt.grid()
plt.show()
```
