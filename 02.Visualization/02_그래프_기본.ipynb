### 그래프 기본

```python
import matplotlib as mpl 
import matplotlib.pyplot as plt 
%matplotlib inline

mpl.rcParams['axes.unicode_minus'] = False   # minus 표시
mpl.rcParams["font.family"] = 'Malgun Gothic'
```

1. Line Plot

```python
x = [0, 1, 2, 3]        # 리스트, 넘파이 어레이 가능, 판다스 시리즈
y = [0, 1, 4, 9]
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, color='m',marker='o', linestyle='--')
plt.grid()
plt.show()
```

```python

import numpy as np 
X = np.arange(4)
Y1 = X * X
Y2 = 1.2 * X * X
Y3 = 0.8 * X * X
plt.plot(X, Y1)
plt.plot(X, Y2, color='m', marker='o', linestyle='--')
plt.plot(X, Y3, 'k^:')
plt.grid()
plt.show()
```

```python
plt.plot(X, Y1, label='Y1')
plt.plot(X, Y2, c='m', marker='o', ls='--', label='Y2')
plt.plot(X, Y3, 'k^:', label='Y3')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c="b",
         lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.title("스타일 적용 예")
plt.show()
```

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c="b",
         lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.title("x축, y축 범위 지정")
plt.xlim(0, 50)
plt.ylim(-5, 20)
plt.show()
```

### 틱 설정

```python
import numpy as np
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("x축과 y축의 tick label 설정")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
plt.yticks([-1, 0, +1])
plt.grid()
plt.show()
```

### x축, y축 라벨, 타이틀

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.plot(X, C, label="cosine")
plt.xlabel("time")
plt.ylabel("amplitude")
plt.title("Cosine Plot")
plt.show()
```
