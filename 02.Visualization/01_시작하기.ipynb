## 데이터 시각화

### Matplotlib 시작하기

```python
import matplotlib.pyplot as plt 
%matplotlib inline
```

### Line plot

```python
plt.plot([1, 4, 9, 16])        # y좌표만 줌. x좌표를 생략하면 [0, 1,, 2, 3]
plt.show()
```

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.title('Line Plot')
plt.grid()
plt.show()
```

```python
import numpy as np
x = np.arange(-4, 4.01, 0.01)
y = np.sin(x)
plt.plot(x, y)
plt.title('Sign Graph')         ## title 한글명으로 하면 오류남
plt.grid()
plt.show()
```
